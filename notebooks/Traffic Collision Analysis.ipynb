```sql
SELECT * FROM (
  SELECT 
    'Peak Month' as metric,
    OCC_MONTH as value,
    COUNT(*) as collisions
  FROM workspace.default.traffic_collisions_open_data_2053198073974531286
  WHERE OCC_YEAR >= 2021
  GROUP BY OCC_MONTH
  ORDER BY collisions DESC
  LIMIT 1
)

UNION ALL

SELECT * FROM (
  SELECT 
    'Peak Day of Week' as metric,
    OCC_DOW as value,
    COUNT(*) as collisions
  FROM workspace.default.traffic_collisions_open_data_2053198073974531286
  WHERE OCC_YEAR >= 2021
  GROUP BY OCC_DOW
  ORDER BY collisions DESC
  LIMIT 1
)

UNION ALL

SELECT * FROM (
  SELECT 
    'Peak Hour' as metric,
    CAST(OCC_HOUR as STRING) as value,
    COUNT(*) as collisions
  FROM workspace.default.traffic_collisions_open_data_2053198073974531286
  WHERE OCC_YEAR >= 2021
  GROUP BY OCC_HOUR
  ORDER BY collisions DESC
  LIMIT 1
)
```

```sql
-- Data Preview: Sample collision records
SELECT 
  EVENT_UNIQUE_ID,
  OCC_DATE,
  OCC_YEAR,
  OCC_MONTH,
  OCC_DOW,
  OCC_HOUR,
  DIVISION,
  NEIGHBOURHOOD_158,
  FATALITIES,
  INJURY_COLLISIONS,
  AUTOMOBILE,
  BICYCLE,
  PEDESTRIAN
FROM workspace.default.traffic_collisions_open_data_2053198073974531286
LIMIT 10
```

```sql
-- Overall Statistics: Dataset summary
SELECT 
  COUNT(*) as total_collisions,
  MIN(OCC_YEAR) as earliest_year,
  MAX(OCC_YEAR) as latest_year,
  COUNT(DISTINCT NEIGHBOURHOOD_158) as num_neighborhoods,
  SUM(FATALITIES) as total_fatalities
FROM workspace.default.traffic_collisions_open_data_2053198073974531286
```

```sql
-- Geographic Data Quality: Validate coordinates and neighborhood data
SELECT 
  COUNT(*) as total_records,
  SUM(CASE WHEN LAT_WGS84 IS NOT NULL AND LONG_WGS84 IS NOT NULL AND LAT_WGS84 != 0.0 AND LONG_WGS84 != 0.0 THEN 1 ELSE 0 END) as valid_coordinates,
  SUM(CASE WHEN NEIGHBOURHOOD_158 IS NOT NULL AND NEIGHBOURHOOD_158 != 'NSA' THEN 1 ELSE 0 END) as valid_neighborhoods,
  COUNT(DISTINCT NEIGHBOURHOOD_158) as unique_neighborhoods
FROM workspace.default.traffic_collisions_open_data_2053198073974531286
```

```sql
-- Peak Hour Analysis: Find hour with most collisions
SELECT 
  OCC_HOUR as hour,
  COUNT(*) as collisions
FROM workspace.default.traffic_collisions_open_data_2053198073974531286
WHERE OCC_YEAR >= 2021
GROUP BY OCC_HOUR
ORDER BY collisions DESC
LIMIT 1
```

```sql
-- Peak Day of Week: Find day with most collisions
SELECT 
  OCC_DOW as day_of_week,
  COUNT(*) as collisions
FROM workspace.default.traffic_collisions_open_data_2053198073974531286
WHERE OCC_YEAR >= 2021
GROUP BY OCC_DOW
ORDER BY collisions DESC
LIMIT 1
```

```sql
-- Peak Month: Find month with most collisions
SELECT 
  OCC_MONTH as month,
  COUNT(*) as collisions
FROM workspace.default.traffic_collisions_open_data_2053198073974531286
WHERE OCC_YEAR >= 2021
GROUP BY OCC_MONTH
ORDER BY collisions DESC
LIMIT 1
```

```sql
-- Vehicle Type Analysis: Distribution of vehicles involved
SELECT 
  CASE 
    WHEN AUTOMOBILE = 'YES' THEN 'Automobile'     
    WHEN BICYCLE = 'YES' THEN 'Bicycle' 
    WHEN PEDESTRIAN = 'YES' THEN 'Pedestrian'     
    WHEN MOTORCYCLE = 'YES' THEN 'Motorcycle' 
    WHEN PASSENGER = 'YES' THEN 'Passenger Vehicle' 
    ELSE 'Other' 
  END AS vehicle_type,
  COUNT(*) as count
FROM workspace.default.traffic_collisions_open_data_2053198073974531286
WHERE OCC_YEAR >= 2021
GROUP BY vehicle_type
ORDER BY count DESC
```
